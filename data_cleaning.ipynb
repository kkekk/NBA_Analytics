```python
# math library
import numpy as np
import pandas as pd
```

```python
college_df = pd.read_csv('college_all.csv')
college_df.isnull().sum() #check for missing data
college_df.columns = ["Player","sos", "year","mp", "ts_pct","efg_pct","fg3a_per_fga_pct", "fta_per_fga_pct","ws_per_40",'g','gs','fg_per_min','fga_per_min','fg_pct','fg2_per_min','fg2a_per_min','fg2_pct','fg3_per_min','fg3a_per_min','fg3_pct','ft_per_min','fta_per_min','ft_pct','trb_per_min','ast_per_min', 'stl_per_min','blk_per_min', 'tov_per_min','pf_per_min','pts_per_min']

college_df = college_df.dropna(how='any').drop(columns = ['gs']) #remove column with many empty cells

print(college_df[college_df.isnull().any(axis=1)].shape) #amount of data to remove
```

```python
college_df.dropna(how='any').to_csv('collegenew.csv') #remove missing data

collegenew = pd.read_csv('collegenew.csv', index_col=0)

print("Are there any rows with empty cells?", collegenew.isnull().values.any() )

collegenew.head()
```

```python
nba_df = pd.read_csv('nba_all.csv', header=None)
nba_df.columns = ['Player','PER 1st', 'PER 2nd', 'Win Shares 1st','Win Shares 2nd', 'Win Shares per 48 min 1st','Win Shares per 48 min 2nd', 'BPM 1st', 'BPM 2nd','VORP 1st', 'VORP 2nd', 'Year Drafted']
nba_df = nba_df.drop(['Year Drafted'], axis=1) #remove duplicate column 
nba_df.head()
```

```python
nba_df.isnull().sum() #check for missing data
```

```python
result = pd.merge(nba_df, college_df, on=['Player', 'Player']) #merge NBA and college data based on player names
result.head()
```

```python

result.dropna(how='any').to_csv('merged_data.csv') #remove missing data
```
