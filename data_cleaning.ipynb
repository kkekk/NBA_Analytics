```python
# math library
import numpy as np
import pandas as pd
```

```python
college_df = pd.read_csv('college_all.csv', header=None)
college_df.isnull().sum() #check for missing data

college_df.columns = ["Player", "Win Shares(Col)", "Win Shares per 40 min(Col)", "Strength of Schedule(Col)", "Year Drafted"]
college_df.head()
```

```python
college_df.dropna(how='any').to_csv('collegenew.csv',header=False) #remove missing data

collegenew = pd.read_csv('collegenew.csv', header=None, index_col=0)

collegenew.columns = ["Player", "Win Shares(Col)", "Win Shares per 40 min(Col)", "Strength of Schedule(Col)", "Year Drafted"]
collegenew.head()
```

```python
collegenew.isnull().sum() #check new data for missing data
```

```python
nba_df = pd.read_csv('nba_all.csv', header=None)
nba_df.columns = ['Player','PER 1st', 'PER 2nd', 'Win Shares 1st','Win Shares 2nd', 'Win Shares per 48 min 1st','Win Shares per 48 min 2nd', 'BPM 1st', 'BPM 2nd','VORP 1st', 'VORP 2nd', 'Year Drafted']
nba_df = nba_df.drop(['Year Drafted'], axis=1) #remove duplicate column 
nba_df.head()
```

```python
nba_df.isnull().sum() #check for missing data
```

```python
result = pd.merge(nba_df, college_df, on=['Player', 'Player']) #merge NBA and college data based on player names
result.head()
```

```python

result.dropna(how='any').to_csv('merged_data.csv') #remove missing data
```
